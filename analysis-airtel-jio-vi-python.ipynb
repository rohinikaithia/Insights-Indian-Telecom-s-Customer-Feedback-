**Finding out # of numerical and categorical columns:**

```python
# list of numerical columns:
numerical_value= [i for i in data.columns if data[i].dtypes != 'O']
print('Number of Numerical columns:', len(numerical_value))
print('')
print('Numerical Variables column:', numerical_value)

#numerical_value: This is a list comprehension that creates a list containing the names of numerical value in the data
#'for i in data.columns': It iterates through each column name in the DataFrame
#if data[i].dtypes != 'O': It checks whether the data type of the values in the current column is not equal to 
#'O'(assuming 'O' represents object/string data type). This condition filters out columns with non-numerical data 
#types
```

```python
# list of categorical columns:
cat_value= [j for j in data.columns if data[j].dtypes == 'O']
print('Number of categorical variables:', len(cat_value))
print("")
print('Categorical Variables column:',cat_value)
```

```python
# Number of unique values in each column
dict={}
for i in data.columns:
    dict[i]=data[i].value_counts().shape[0]
pd.DataFrame(dict,index=['Unique']).T.style.background_gradient(cmap='bwr')
```

```python
#Visualizing Missing Values
import missingno as msno
msno.matrix(data,figsize=(10,4))
```

```python
#Percentage of Missing Values

def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    Percentage = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, Percentage], axis=1, keys=['Total', 'Percentage'])
missing_data(data).style.background_gradient(cmap='coolwarm')
```
